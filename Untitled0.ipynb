```python
#importar dados

import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja_1 = pd.read_csv(url)
loja_2 = pd.read_csv(url2)
loja_3 = pd.read_csv(url3)
loja_4 = pd.read_csv(url4)

# Adiciona a coluna 'Loja' a cada DataFrame individualmente
loja_1['Loja'] = 'Loja 1'
loja_2['Loja'] = 'Loja 2'
loja_3['Loja'] = 'Loja 3'
loja_4['Loja'] = 'Loja 4'

# Concatena os DataFrames após adicionar a coluna 'Loja'
todas_lojas = pd.concat([loja_1, loja_2, loja_3, loja_4])
todas_lojas.shape
todas_lojas['Preço'].sum()

todas_lojas.groupby('Loja')['Preço'].sum()

todas_lojas.head()

todas_lojas = pd.concat([loja_1, loja_2, loja_3, loja_4])
todas_lojas['Loja'].value_counts()

todas_lojas.groupby('Produto')['Preço'].sum()

todas_lojas.groupby('Produto')['Preço'].sum().sort_values(ascending=False)

todas_lojas.groupby('Categoria do Produto')['Preço'].sum()
tabela_categoria = todas_lojas.pivot_table(
    values='Preço',
    index='Loja',
    columns='Categoria do Produto',
    aggfunc='sum'
)

tabela_categoria
faturamento_por_loja = (
    todas_lojas
    .groupby('Loja', as_index=False)['Preço']
    .sum()
)

faturamento_por_loja

#faturamento total de cada loja





```

```python
#Vendas por Categoria

tabela_categoria = todas_lojas.pivot_table(
    values='Preço',
    index='Loja',
    columns='Categoria do Produto',
    aggfunc='sum'
)

tabela_categoria

```

```python
#media de avaliacao das lojas

media_avaliacao = (
    todas_lojas
    .groupby('Loja', as_index=False)['Avaliação da compra']
    .mean()
)

media_avaliacao


```

```python
import matplotlib.pyplot as plt

faturamento = todas_lojas.groupby('Loja')['Preço'].sum()

faturamento.plot(kind='bar')
plt.title('Faturamento por Loja')
plt.ylabel('Faturamento')
plt.xlabel('Loja')
plt.xticks(rotation=0)
plt.show()
```

```python
#Produtos Mais e Menos Vendidos
produtos_faturamento = (
    todas_lojas
    .groupby('Produto')['Preço']
    .sum()
    .sort_values(ascending=False)
)

produtos_faturamento

```

```python
produtos_qtd = todas_lojas['Produto'].value_counts()
top10 = produtos_qtd.head(10)

top10.plot(kind='bar')
plt.title('Top 10 Produtos Mais Vendidos')
plt.ylabel('Quantidade de Vendas')
plt.xticks(rotation=45)
plt.show()
```

```python
menos_vendidos = produtos_qtd.tail(10)

menos_vendidos.plot(kind='bar')
plt.title('Top 10 Produtos Menos Vendidos')
plt.ylabel('Quantidade de Vendas')
plt.xticks(rotation=45)
plt.show()
```
